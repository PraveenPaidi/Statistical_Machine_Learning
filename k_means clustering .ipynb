```python
from mnist import MNIST
import numpy as np
import math
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
mndata=MNIST(r'C:\Users\ppaidi')
X_train, labels_train = map(np.array, mndata.load_training())
X_test, labels_test = map(np.array, mndata.load_testing())
X_train = X_train/255.0
X_test = X_test/255.0
(n,d)=np.shape(X_train)
(n1,d)=np.shape(X_test)

trainerror=[]
testerror=[]

k1=[2, 5, 10, 20, 40, 80, 160, 320, 640, 1280]
for z in range(10):
    k=k1[z]
    if z>0:
        trainerror.append(c)
        testerror.append(c1)
    
    #initializing random centroid matrix
    intt=np.random.randint(0,n-1, size=k).flatten()
    mu=X_train[intt].T
    
    diff=np.zeros((n,k))
    difftest=np.zeros((n1,k))

    for j in range(30):
        for i in range(k):
            mu1=mu[:,i].reshape(d,1)
            temp=np.dot(mu1,np.ones((1,n))).T
            diff[:,i]=np.linalg.norm((X_train-temp),ord=2,axis=1)
            temp=np.dot(mu1,np.ones((1,n1))).T
            difftest[:,i]=np.linalg.norm((X_test-temp),ord=2,axis=1)

        #assigning cluster and updating mean
        cluster=np.argmin(diff,axis=1)
        
        #training error
        c=np.mean(np.min(diff,axis=1),axis=0)
        
        #testing error
        c1=np.mean(np.min(difftest,axis=1),axis=0)
        
        for i in range(k):
            indices=np.argwhere(cluster == i).flatten()
            Xtemp=X_train[indices]
            mu[:,i]=np.mean(Xtemp,axis=0)
                
trainerror.append(c)
testerror.append(c1)               
plt.plot(k1,c,label='Training error')
plt.plot(k1,c1,label='Test Error')
plt.xlabel('K values')
plt.ylabel('error')
plt.title('Training error and Test error vs k values')
plt.legend()
plt.show()
```
